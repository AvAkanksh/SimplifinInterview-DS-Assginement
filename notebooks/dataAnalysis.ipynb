# Duplicate Bank_Transactions in Paytm_data 
- Upon analysis I have found that there are duplicate Bank_Transaction_ID in paytm_data but mostly they are on different date so that wont be a problem the below code checks that 
  

```python
duplicate_bank_transaction_ids = set(paytm_data[paytm_data.duplicated(subset=['Bank_Transaction_ID'])]['Bank_Transaction_ID'])
```

```python
for bank_transaction_id in duplicate_bank_transaction_ids:
    matched = paytm_data[paytm_data['Bank_Transaction_ID'] == bank_transaction_id]
    matched.reset_index(inplace=True)
    first_match_date = str(matched['Standardized DateTime'][0])[:8]
    for j in range(1,len(matched)):
        if(str(matched['Standardized DateTime'][j])[:8] == first_match_date):
            print("They have same date for Transaction ID : ",bank_transaction_id)
```

### 4652 records in Paytm_EDC and 6680 records in OMS.

- so we will try to match each record from Paytm_EDC with OMS. 
- The reason for having more records in OMS is because there are many ZERO bills generated.
- Also some of the payments are done by cash so those bills won't be present in Paytm_EDC records.
  

```python
paytm_data["Matched"] = False
paytm_data["Receipt #"] = ''
paytm_data["Discrepancy"] = ''
paytm_data["MatchingTechnique"] = ''
oms_data["Matched"] = False
oms_data["Transaction_ID"] = ''
oms_data["Discrepancy"] = ''
oms_data["MatchingTechnique"] = ''
```

# Direct Matching 
### (using Bank Transcation ID in Paytm EDC with Payment Aggregator Transaction ID in OMS
 

```python
matches_from_direct_matching = 0

for i in range(len(paytm_data)):
    match = oms_data[str(paytm_data['Bank_Transaction_ID'][i]) == oms_data['Payment Aggregator Transaction ID']]
    if not match.empty:
        # match.reset_index(inplace=True)
        for j, row in match.iterrows():
            if(str(match['Standardized DateTime'][j])[:8] == str(paytm_data['Standardized DateTime'][i])[:8]):
                paytm_data.loc[i,'Matched'] = True
                paytm_data.loc[i,'Receipt #'] = match['Receipt #'][j]
                paytm_data.loc[i,'MatchingTechnique'] = 'Direct'
                oms_data.loc[j,'Matched'] = True
                oms_data.loc[j,'Transaction_ID'] = paytm_data.loc[i,'Transaction_ID']
                oms_data.loc[j,'MatchingTechnique'] = 'Direct'
                matches_from_direct_matching+=1
                # All the payments in Paytm EDC are made by CreditCard / Debit card
                # Therefore, we can use the "Settled_Amount" column to check if the amounts match with the "Card" column amount in oms_data

                if(match['Card'][j] != paytm_data['Settled_Amount'][i]):
                    paytm_data['Discrepancy'][i] = "The amounts didn't match for Transaction ID : ",paytm_data['Bank_Transaction_ID'][i] , " The amounts are 'Amount after Advance/Package' : ",match['Amount after Advance/Package'][j]," Amount Due : ",match['Amount Due'][j] , " Settled Amount : ", paytm_data['Settled_Amount'][i]
                break

print("Matches from Direct Matching Technique : ",matches_from_direct_matching)
```

# Fuzzy Matching
### (using "UnixTimestampAmount" column in Paytm EDC with "UnixTimestampAmount" column in OMS)

```python
# Records which are not matched yet after Direct Matching Technique is applied
paytm_data_unmatched_records = paytm_data[~paytm_data['Matched']]
```

```python
# Logic for Fuzzy matching

# Technique 1 : Levenstein Distance

# I have read this post for understanding how to use thefuzz library and which score to use when
# https://stackoverflow.com/questions/31806695/when-to-use-which-fuzz-function-to-compare-2-strings

count = 0
for index , row in paytm_data_unmatched_records.iterrows():
    # perform fuzzy matching on UnixTimestampAmount columns
    possible_matches = process.extractOne(row['UnixTimestampAmount'],oms_data['UnixTimestampAmount'], scorer=fuzz.token_sort_ratio)
    print(f"""
          paytm_data (Date) : {row["Standardized DateTime"]}
          paytm_data (Amount) : {row["UnixTimestampAmount"]}
          oms_data (Date) : {oms_data.loc[possible_matches[2],"Standardized DateTime"]}
          oms_data (Date Amount) : {oms_data.loc[possible_matches[2],"UnixTimestampAmount"]}
          """)
    if(count >= 10):
        break
    count+=1
    
```

- Modifying the search space, by limiting the search only on the records which are close by in time. 

### UnixTimestampAmount

```python
threshold = 90 # in seconds
fuzzySearchColumn = "UnixTimestampAmount"
count = 0
unique_ids = set()

for index , row in paytm_data_unmatched_records.iterrows():
    # Performing fuzzy matching on UnixTimestampAmount column and limit the search space only on the records which are close by in time based on the Threshold value and matching amounts only.
    possible_matches = process.extract(row[fuzzySearchColumn],oms_data[(abs(row['UnixTimestamp']-oms_data['UnixTimestamp'])<=threshold ) & (row["FinalAmount"]==oms_data['FinalAmount']) ][fuzzySearchColumn], scorer=fuzz.token_sort_ratio)
    if len(possible_matches):
        print(row[fuzzySearchColumn])
        print(possible_matches)
        count+=1
        
        for possible_match in possible_matches:
            unique_ids.add(possible_match[2])

print("Total Fuzzy Matches Found : ",count)
print("Unique IDs : ",len(unique_ids))
```
